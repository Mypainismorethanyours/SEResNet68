## Experiment with/without Data Augmentation
### Data Augmentation provides a better performance and is used in our final model.

```python
import matplotlib.pyplot as plt
import pandas as pd

with_aug = pd.read_csv('./batch_size32_model/train_test_eval.csv')
without_aug = pd.read_csv('./without_aug_model/train_test_eval.csv')
```

```python
fig_loss, ax_loss = plt.subplots()

ax_loss.set_xlabel('Epoch')
ax_loss.set_ylabel('Loss')
ax_loss.plot(with_aug.index, with_aug['Train Loss'], label='Train Loss with data augmentation')
ax_loss.plot(with_aug.index, with_aug['Test Loss'], label='Test Loss with data augmentation', linestyle='dashed')
ax_loss.plot(without_aug.index, without_aug['Train Loss'], label='Train Loss without data augmentation')
ax_loss.plot(without_aug.index, without_aug['Test Loss'], label='Test Loss without data augmentation', linestyle='dashed')
ax_loss.set_title('Training and Testing Loss')
ax_loss.legend()

plt.savefig('./plots/augmentation_loss')

fig_acc, ax_acc = plt.subplots()

ax_acc.set_xlabel('Epoch')
ax_acc.set_ylabel('Accuracy')
ax_acc.plot(with_aug.index, with_aug['Train Accuracy'], label='Train Accuracy with data augmentation')
ax_acc.plot(with_aug.index, with_aug['Test Accuracy'], label='Test Accuracy with data augmentation', linestyle='dashed')
ax_acc.plot(without_aug.index, without_aug['Train Accuracy'], label='Train Accuracy without data augmentation')
ax_acc.plot(without_aug.index, without_aug['Test Accuracy'], label='Test Accuracy without data augmentation', linestyle='dashed')
ax_acc.set_title('Training and Testing Accuracy')
ax_acc.legend()

plt.savefig('./plots/augmentation_acc')

plt.show()
```
