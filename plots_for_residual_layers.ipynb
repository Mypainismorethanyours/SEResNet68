## Experiment on Different ResNet Architecture: 3 Residual Layers VS 4 Residual Layers
### ResNet with 3 residual layers and hyperparameters N = 3, C = 64, Bi = [4, 4, 3], Fi = [3, 3, 3], Ki = [1, 1, 1], P = 8 provides a better performance and is used in our final model.

```python
import matplotlib.pyplot as plt
import pandas as pd

three_residual_layers = pd.read_csv('./batch_size32_model/train_test_eval.csv')
four_residual_layers = pd.read_csv('./4residual_layers_model/train_test_eval.csv')
```

```python
fig_loss, ax_loss = plt.subplots()

ax_loss.set_xlabel('Epoch')
ax_loss.set_ylabel('Loss')
ax_loss.plot(three_residual_layers.index, three_residual_layers['Train Loss'], label='Train Loss with 3 residual layers[4,4,3]')
ax_loss.plot(three_residual_layers.index, three_residual_layers['Test Loss'], label='Test Loss with 3 residual layers[4,4,3]', linestyle='dashed')
ax_loss.plot(four_residual_layers.index, four_residual_layers['Train Loss'], label='Train Loss with 4 residual layers[2,2,2,2]')
ax_loss.plot(four_residual_layers.index, four_residual_layers['Test Loss'], label='Test Loss with 4 residual layers[2,2,2,2]', linestyle='dashed')
ax_loss.set_title('Training and Testing Loss')
ax_loss.legend()

plt.savefig('./plots/redidual_layers(num_blocks)_loss')

fig_acc, ax_acc = plt.subplots()

ax_acc.set_xlabel('Epoch')
ax_acc.set_ylabel('Accuracy')
ax_acc.plot(three_residual_layers.index, three_residual_layers['Train Accuracy'], label='Train Accuracy with 3 residual layers[4,4,3]')
ax_acc.plot(three_residual_layers.index, three_residual_layers['Test Accuracy'], label='Test Accuracy with 3 residual layers[4,4,3]', linestyle='dashed')
ax_acc.plot(four_residual_layers.index, four_residual_layers['Train Accuracy'], label='Train Accuracy with 4 residual layers[2,2,2,2]')
ax_acc.plot(four_residual_layers.index, four_residual_layers['Test Accuracy'], label='Test Accuracy with 4 residual layers[2,2,2,2]', linestyle='dashed')
ax_acc.set_title('Training and Testing Accuracy')
ax_acc.legend()

plt.savefig('./plots/redidual_layers(num_blocks)_acc')

plt.show()
```
