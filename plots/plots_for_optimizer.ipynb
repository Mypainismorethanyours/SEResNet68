## Experiment on Different Optimizers: Lookahead+SGD, Ranger (Lookahead+RAdam)
### Ranger provides a better performance and is used in our final model.

```python
import matplotlib.pyplot as plt
import pandas as pd

lookahead_sgd = pd.read_csv('./batch_size32_model/train_test_eval.csv')
ranger = pd.read_csv('./ranger_model/train_test_eval.csv')
```

```python
fig_loss, ax_loss = plt.subplots()

ax_loss.set_xlabel('Epoch')
ax_loss.set_ylabel('Loss')
ax_loss.plot(lookahead_sgd.index, lookahead_sgd['Train Loss'], label='Train Loss with Lookahead+SGD')
ax_loss.plot(lookahead_sgd.index, lookahead_sgd['Test Loss'], label='Test Loss with Lookahead+SGD', linestyle='dashed')
ax_loss.plot(ranger.index, ranger['Train Loss'], label='Train Loss with Ranger(Lookahead+RAdam)')
ax_loss.plot(ranger.index, ranger['Test Loss'], label='Test Loss with Ranger(Lookahead+RAdam)', linestyle='dashed')
ax_loss.set_title('Training and Testing Loss')
ax_loss.legend()

plt.savefig('./plots/optimizer_loss')

fig_acc, ax_acc = plt.subplots()

ax_acc.set_xlabel('Epoch')
ax_acc.set_ylabel('Accuracy')
ax_acc.plot(lookahead_sgd.index, lookahead_sgd['Train Accuracy'], label='Train Accuracy with Lookahead+SGD')
ax_acc.plot(lookahead_sgd.index, lookahead_sgd['Test Accuracy'], label='Test Accuracy with Lookahead+SGD', linestyle='dashed')
ax_acc.plot(ranger.index, ranger['Train Accuracy'], label='Train Accuracy with Ranger(Lookahead+RAdam)')
ax_acc.plot(ranger.index, ranger['Test Accuracy'], label='Test Accuracy with Ranger(Lookahead+RAdam)', linestyle='dashed')
ax_acc.set_title('Training and Testing Accuracy')
ax_acc.legend()

plt.savefig('./plots/optimizer_acc')

plt.show()
```
