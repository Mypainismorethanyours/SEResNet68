## Experiment on lr = 0.1 and lr = 0.01
### lr = 0.1 provides a better performance and is used in our final model.

```python
import matplotlib.pyplot as plt
import pandas as pd

lr_0_1 = pd.read_csv('./batch_size32_model/train_test_eval.csv')
lr_0_01 = pd.read_csv('./lr0.01_model/train_test_eval.csv')
```

```python
fig_loss, ax_loss = plt.subplots()

ax_loss.set_xlabel('Epoch')
ax_loss.set_ylabel('Loss')
ax_loss.plot(lr_0_1.index, lr_0_1['Train Loss'], label='Train Loss with lr=0.1')
ax_loss.plot(lr_0_1.index, lr_0_1['Test Loss'], label='Test Loss with lr=0.1', linestyle='dashed')
ax_loss.plot(lr_0_01.index, lr_0_01['Train Loss'], label='Train Loss with lr=0.01')
ax_loss.plot(lr_0_01.index, lr_0_01['Test Loss'], label='Test Loss with lr=0.01', linestyle='dashed')
ax_loss.set_title('Training and Testing Loss')
ax_loss.legend()

plt.savefig('./plots/optimizer_loss')

fig_acc, ax_acc = plt.subplots()

ax_acc.set_xlabel('Epoch')
ax_acc.set_ylabel('Accuracy')
ax_acc.plot(lr_0_1.index, lr_0_1['Train Accuracy'], label='Train Accuracy with lr=0.1')
ax_acc.plot(lr_0_1.index, lr_0_1['Test Accuracy'], label='Test Accuracy with lr=0.1', linestyle='dashed')
ax_acc.plot(lr_0_01.index, lr_0_01['Train Accuracy'], label='Train Accuracy with lr=0.01')
ax_acc.plot(lr_0_01.index, lr_0_01['Test Accuracy'], label='Test Accuracy with lr=0.01', linestyle='dashed')
ax_acc.set_title('Training and Testing Accuracy')
ax_acc.legend()

plt.savefig('./plots/optimizer_acc')

plt.show()
```
