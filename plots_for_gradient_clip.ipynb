```python
import matplotlib.pyplot as plt
import pandas as pd

with_gradient_clip = pd.read_csv('./batch_size32_model/train_test_eval.csv')
without_gradient_clip = pd.read_csv('./without_gradient_model/train_test_eval.csv')
```

```python
fig_loss, ax_loss = plt.subplots()

ax_loss.set_xlabel('Epoch')
ax_loss.set_ylabel('Loss')
ax_loss.plot(with_gradient_clip.index, with_gradient_clip['Train Loss'], label='Train Loss with gradient clip=0.1')
ax_loss.plot(with_gradient_clip.index, with_gradient_clip['Test Loss'], label='Test Loss with gradient clip=0.1', linestyle='dashed')
ax_loss.plot(without_gradient_clip.index, without_gradient_clip['Train Loss'], label='Train Loss without gradient clip')
ax_loss.plot(without_gradient_clip.index, without_gradient_clip['Test Loss'], label='Test Loss without gradient clip', linestyle='dashed')
ax_loss.set_title('Training and Testing Loss')
ax_loss.legend()

plt.savefig('./plots/gradient_clip_loss')

fig_acc, ax_acc = plt.subplots()

ax_acc.set_xlabel('Epoch')
ax_acc.set_ylabel('Accuracy')
ax_acc.plot(with_gradient_clip.index, with_gradient_clip['Train Accuracy'], label='Train Accuracy with gradient clip=0.1')
ax_acc.plot(with_gradient_clip.index, with_gradient_clip['Test Accuracy'], label='Test Accuracy with gradient clip=0.1', linestyle='dashed')
ax_acc.plot(without_gradient_clip.index, without_gradient_clip['Train Accuracy'], label='Train Accuracy without gradient clip')
ax_acc.plot(without_gradient_clip.index, without_gradient_clip['Test Accuracy'], label='Test Accuracy without gradient clip', linestyle='dashed')
ax_acc.set_title('Training and Testing Accuracy')
ax_acc.legend()

plt.savefig('./plots/gradient_clip_acc')

plt.show()
```
